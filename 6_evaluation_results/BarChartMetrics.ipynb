```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

retrievers = ['BM25', 'OpenAI Embedding', 'SentenceTransformers', 'German DPR']
metrics = ['Recall', 'Precision', 'MAP', 'MRR', 'NDCG']

data = np.array([
    [1.0000, 0.2000, 0.9762, 0.9762, 0.9824],
    [1.0000, 0.2000, 0.9735, 0.9735, 0.9803],
    [0.8571, 0.1714, 0.6894, 0.6894, 0.7306],
    [0.6032, 0.1206, 0.4421, 0.4421, 0.4820]
])

df = pd.DataFrame(data, columns=metrics)
df['Retrievers'] = retrievers

df_melted = pd.melt(df, id_vars='Retrievers', var_name='Metrics', value_name='Scores')

sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.barplot(x='Retrievers', y='Scores', hue='Metrics', data=df_melted, palette='colorblind')

plt.xlabel('Retrievers')
plt.ylabel('Scores')
plt.title('Comparison of Metrics for FAQ Dataset')

plt.show()
```
